Nodal Analysis

Pe - delP_r = Pwf(IPR)

Pwf(TPR)  =  P_sep + delP_t + delp_sl

Pe - delP_r =  P_sep + delP_t + delp_sl (Intersection at unique pressure)

```python
#import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
q_ipr = [0,1999,3094,4512,4963,5257,5458,5519]
pwf_ipr = [4000,3500,3000,2500,2000,1500,1000,500,14.7]

ipr_df= pd.DataFrame({'pwf':[4000,3500,3000,2500,2000,1500,1000,500,14.7],
                      'Q':[0,1999,3094,3902,4512,4963,5257,5458,5519]})

ipr = {'pwf':pwf_ipr,
       'Q': q_ipr}          #we Have created a dictionary

ipr_df
```

```python

x_ipr = ipr_df['Q']
y_ipr = ipr_df['pwf']

plt.plot(x_ipr,y_ipr)
plt.xlabel('Flow Rate (Q)')
plt.ylabel('Bottom-hole Flowing Pressure (Pwf)')
plt.title('Inflow Performance Relationship (IPR)')
plt.grid(True)
plt.show()
```

```python
#Now plotting the TPR


p_190 = [1334,1400,1487,1592,1712,1843,1984,2132,2287,2446,2609]
p_2375 =[1298,1320,1351,1390,1435,1487,1545,1609,1677,1749,1824]
p_2875 = [1286,1294,1305,1319,1336,1356,1378,1403,1431,1461,1493]

#Creating Dataframe
tpr=pd.DataFrame({"Q":q_tpr,
                  "Pwf(1.9 tbg)":p_190,
                  "Pwf(2.375 tbg)":p_2375,
                  "Pwf(2.875 tbg)":p_2875})
tpr



```

```python
#Plotting Tubing Performance Relation (TPR) Curve
plt.figure(figsize=(12,6))
plt.plot(tpr["Q"],tpr["Pwf(1.9 tbg)"], label="TPR for 1.9 tubing ",linewidth=3)
plt.plot(tpr["Q"],tpr["Pwf(2.375 tbg)"], label="TPR for 2.375 tubing ",linewidth=3)
plt.plot(tpr["Q"],tpr["Pwf(2.875 tbg)"], label="TPR for 2.875 tubing ",linewidth=3)
plt.xlabel(" Q (Mscf/d)")
plt.ylabel(" Pwf (psia)")
plt.xlim(0,6500)
plt.ylim(0,3000)
plt.title("Tubing Performance Relation (TPR) Curve")
plt.grid(True)
plt.legend(loc="best")
```

```python
#Plotting IPR and TPR
plt.figure(figsize=(12,8))
plt.style.use("classic")
plt.plot(ipr_df["Q"],ipr_df["pwf"],label="IPR Curve",lw=5)
plt.plot(tpr["Q"],tpr["Pwf(1.9 tbg)"], label="TPR for 1.9 tubing ",linewidth=3)
plt.plot(tpr["Q"],tpr["Pwf(2.375 tbg)"], label="TPR for 2.375 tubing ",linewidth=3)
plt.plot(tpr["Q"],tpr["Pwf(2.875 tbg)"], label="TPR for 2.875 tubing ",linewidth=3)
plt.xlim(0,6500)
plt.ylim(0,4500)
plt.xlabel(" Q (Mscf/d)")
plt.ylabel(" Pwf (psia)")
plt.title("Plot of IPR and TPR Curve ")
plt.grid(True)
plt.legend()
```
